````python
class Document:
    def __init__(self, page_content, metadata):
        self.page_content = page_content
        self.metadata = metadata

def dataframe_to_qdrant_documents(df):
    """
    Converts a DataFrame with specific columns into a list of Document objects
    suitable for creating a Qdrant collection.

    Args:
        df (pandas.DataFrame): The DataFrame to convert.

    Returns:
        list: A list of Document objects, where each object represents a row in
              the DataFrame and adheres to Qdrant's document structure.
    """

    documents = []
    for index, row in df.iterrows():
        metadata = {col: row[col] for col in df.columns if col not in ['use_cases']}
        documents.append(Document(
            page_content=row["use_cases"],
            metadata=metadata,
        ))
    return documents

# Assuming your DataFrame is named 'strategies'
qdrant_documents = dataframe_to_qdrant_documents(strategies)

url="https://f4bf0d03-f13c-43b6-87cc-32935393ab68.us-east4-0.gcp.cloud.qdrant.io:6333"
api_key="r8CbzHB7iEGgX5ISK885jtMkaHCLJYAceYvjx3V-OEtKHqn1xnzHvQ"

qdrant = Qdrant.from_documents(
    qdrant_documents,  # Use the list of Document objects suitable for Qdrant
    embeddings,
    url=url,
    prefer_grpc=True,
    api_key=api_key,
    collection_name="biomimicry-strategies-new",
    content_payload_key="page_content",
    metadata_payload_key="metadata",
)
````
